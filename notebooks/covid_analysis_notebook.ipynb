```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the COVID-19 dataset
df_covid = pd.read_csv('../data/owid-covid-data.csv')
```

```python
# Display the first five rows of the dataset (only works in notebooks)
df_covid.head(5)
```

```python
# Display the shape of the dataset (rows, columns)
df_covid.shape
```

```python
# Display the column names in the dataset
df_covid.columns
```

```python
# Check for missing values in the dataset
# This will return a DataFrame of True/False for each cell
missing = df_covid.isnull()
missing
```

```python
# Display data types of all columns
df_covid.dtypes
```

```python
# Convert 'date' column to datetime format for time-based analysis
df_covid['date'] = pd.to_datetime(df_covid['date'])
```

```python
# Select only the necessary columns for analysis
columns_needed = [
    'location', 'date', 'total_cases', 'new_cases', 'total_deaths', 
    'new_deaths', 'people_vaccinated', 'people_fully_vaccinated', 
    'population'
]

df_covid = df_covid[columns_needed]
df_covid
```

```python
# Replace missing (NaN) values with 0 to simplify analysis
df_covid = df_covid.fillna(0)

# Save the cleaned dataset to a new CSV file
df_covid.to_csv('../output/cleaned_covid_data.csv')
```

```python
# Group the dataset by country ('location') and get the max total_cases per country
d = df_covid.groupby('location')['total_cases'].max().reset_index()

# Sort the countries in descending order based on total_cases
top_ten_countries_df = d.sort_values(by='total_cases', ascending=False)

# Define a list of non-country regions to exclude from top 10
mask = ['World', 'High-income countries', 'Asia', 'Europe', 'Upper-middle-income countries', 
        'European Union (27)', 'North America', 'Lower-middle-income countries', 'South America']

# Filter out the non-country regions and select the top 10 countries
top_10_clean = top_ten_countries_df[~top_ten_countries_df['location'].isin(mask)].head(10)

# Display the top 10 countries with highest total cases
top_10_clean
```

```python
# Extract the list of top 10 countries
top_10_list = top_10_clean['location'].values.tolist()

# Filter the original dataframe to include only records from these top 10 countries
final_df = df_covid[df_covid['location'].isin(top_10_list)]
final_df
```

```python
# Plot total COVID-19 cases over time for the top 10 countries
plt.figure(figsize=(12,6))
sns.lineplot(data=final_df, x='date', y='total_cases', hue = 'location')
plt.title("Total COVID-19 Cases Over Time for Selected Countries")
plt.xlabel("Date")
plt.ylabel("Total Cases")
plt.show()
```

```python
# Plot total deaths for each of the top 10 countries
plt.figure(figsize = (15,6))
sns.barplot(data = final_df, x = 'location', y = 'total_deaths')
plt.title("Total deaths of each country")
plt.xlabel("counties")
plt.ylabel("Total deaths")
plt.show()
```

```python
# Plot total number of people vaccinated over time (all countries combined)
plt.figure(figsize = (15,6))
sns.barplot(data = final_df, x = 'location', y = 'people_vaccinated')
plt.title("Total vaccinated over time")
plt.xlabel("counties")
plt.ylabel("Total vaccinated")
plt.show()
```
